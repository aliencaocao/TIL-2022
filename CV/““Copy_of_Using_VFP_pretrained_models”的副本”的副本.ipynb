```python
# NOT NEEDED TO RUN, RUN CELL ABOVE ONLY

!mkdir "/content/drive/MyDrive/TIL 2022/CV/preds/retinanet/"

import json
import os
import numpy as np
from pathlib import Path
from mmdet.apis import init_detector, inference_detector
import mmcv
from tqdm import tqdm

config_file = '/content/VFP290K/configs/VFP290K/retinanet_r50_1x_benchmark.py'
checkpoint_file = '/content/drive/MyDrive/TIL 2022/CV/retinanet_training/epoch_52.pth'

model = init_detector(config_file, checkpoint_file, device='cuda:0')

with open("/content/CV_data/test/interim_no_annotations.json") as img_info_file:
  img_info = json.load(img_info_file)["images"]
filename_to_id = {img["file_name"]: img["id"] for img in img_info}

preds = []
confidence_threshold = 0.3
for i, img_path in enumerate(tqdm(list(Path("/content/CV_data/test/images").glob("*")))):
  person_detections = inference_detector(model, img_path)[0]
  detections_filtered = person_detections[np.where(person_detections[:, 4] > confidence_threshold)]

  img_id = filename_to_id[os.path.basename(img_path)]
  for detection in detections_filtered:
    x1, y1, x2, y2, confidence = [float(x) for x in detection] # convert from np.float32 to float
    preds.append({
      "image_id": img_id,
      "bbox": [x1, y1, x2-x1, y2-y1],
      "category_id": 1, # person
      "score": confidence,
    })

with open("/content/drive/MyDrive/TIL 2022/CV/preds/retinanet/retinanet_epoch52.json", "w") as preds_file:
  json.dump(preds, preds_file, separators=(",", ":"))
```
